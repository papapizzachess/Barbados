```python
import pandas as pd

# Load CSVs
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test_with_path.csv")

# Keep only needed columns
train_df = train_df[["ID", "geometry"]]
test_df = test_df[["ID", "image_path"]]

# Drop duplicate IDs (keep first occurrence)
train_df = train_df.drop_duplicates(subset="ID", keep="first")
test_df = test_df.drop_duplicates(subset="ID", keep="first")

# Merge on ID
merged_df = pd.merge(
    test_df,
    train_df,
    on="ID",
    how="inner"
)

# Ensure one row per ID after merge
merged_df = merged_df.drop_duplicates(subset="ID", keep="first")

# Save result
merged_df.to_csv("geo_poly.csv", index=False)

print(f"Saved {len(merged_df)} unique IDs to merged.csv")
```
